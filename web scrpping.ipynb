## web scraping converting into csv file

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.sunbeaminfo.com/placements-record-details.php?batch_name=DBDA-Sep-2023&percentage=60"
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table=soup.find_all('table',class_="table table-bordered table-striped")[0]
print(table)
```

```python
title_list=table.find_all('th')
print(title_list)
```

```python
title=[t.text for t in title_list]
print(title)
```

```python
df=pd.DataFrame(columns=title)
df
```

```python
row_list=table.find_all('tr')[1:]
print(row_list)
```

```python
data_list=[]
for row in row_list:
    data=row.find_all('td')
    data_list.append([d.text for d in data])

print(data_list)
    
```

```python
df=pd.DataFrame(data_list,columns=df.columns)
```

```python
df
```

```python
df.to_csv("sunbeam dbda placement sept 2023 1.csv",index=False)
```
